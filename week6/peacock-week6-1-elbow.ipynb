```python
#Load Iris Dataset and initialize it
from sklearn.datasets import load_iris

iris = load_iris()

print(iris.DESCR)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head()
```

```python
from sklearn.cluster import KMeans

X = df.iloc[:,[0,1,2,3]].values
wcss = []

#checking different k values and storing the results in wcss
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k,random_state=11)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss, color="red")
plt.title("Elbow Method to find Optimal K for K-Means Cluster")
plt.xlabel("Number of Clusters")
plt.xticks(range(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

<p>According to the elbow method, the optimal K value will be 3</p>
